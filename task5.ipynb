```python
x=df[["TV","Radio","Newspaper"]]
y=df.Sales.values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mod=LinearRegression()
```

```python
mod.fit(x_train,y_train)
```

```python
y_pred=mod.predict(x_test)
```

```python
print("The mean squared error : ",mean_squared_error(y_test,y_pred))
```

```python
print("The absolute error : ",mean_absolute_error(y_test,y_pred))
```

```python
data_1 = pd.DataFrame({'TV': [50], 'Radio': [80], 'Newspaper': [10]})
predicted_sales_1 = mod.predict(data_1)
print("Predicted Sales (Data 1):", predicted_sales_1)
```
